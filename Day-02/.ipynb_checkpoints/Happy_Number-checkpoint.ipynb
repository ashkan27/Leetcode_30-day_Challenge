## Happy Number

### Problem Statement:

Write an algorithm to determine if a number is "happy".

A happy number is a number defined by the following process: Starting with any positive integer, replace the number by the sum of the squares of its digits, and repeat the process until the number equals 1 (where it will stay), or it loops endlessly in a cycle which does not include 1. Those numbers for which this process ends in 1 are happy numbers.

### Code:

```python
def isHappy(n):
    h=set()
    h.add(n)
    while(True):
        s = 0
        for i in str(n):
            s += pow(int(i),2)
        if s == 1:
            return True
        if s in h:
            return False
        h.add(s)
        n = s
```

### Testing:

#### Example 1:

```python
isHappy(19)
```

Input: 19

Output: True

Explanation:

1^2 + 9^2 = 82 ||
8^2 + 2^2 = 68 ||
6^2 + 8^2 = 100 ||
1^2 + 0^2 + 0^2 = 1

#### Example 2:

```python
isHappy(2)
```

Input: 2

Output: False

Explanation:

2^2 = 4 || 4^2 = 16 || 1^2 + 6^2 = 37 || 3^2 + 7^2 = 58 || 5^2 + 8^2 = 89 || 8^2 + 9^2 = 145 || 1^2 + 4^2 + 5^2 = 42 || 4^2 + 2^2 = 20 || 2^2 + 0^2 = 4 (Endless Loop)
