## Move Zeroes

### Problem Statement:

Given an array nums, write a function to move all 0's to the end of it while maintaining the relative order of the non-zero elements.

### Code:

```python
def moveZeroes(nums):
    j = 0
    for i in range(len(nums)-1):
        if nums[i-j] == 0:
            nums.pop(i-j)
            j += 1
            nums.append(0)
    return nums
```

### Testing:

#### Example 1:

```python
nums = [0,0,1,2,1,0,3,7,0,2]
moveZeroes(nums)
```

#### Example 2:

```python
nums = [0,0,0,0,0,7,2,0,1,0,0,2,0]
moveZeroes(nums)
```
