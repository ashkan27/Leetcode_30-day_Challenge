## Minimum Path Sum

### Problem Statement:

Given a m x n grid filled with non-negative numbers, find a path from top left to bottom right which minimizes the sum of all numbers along its path.

Note: You can only move either down or right at any point in time.

### Code:

```python
def minPathSum(grid):
    n = len(grid)
    if n == 0:
        return 0
    m = len(grid[0]) 
    ans = []
    ans.append([grid[0][0]])
    for i in range(1, n): 
        ans.append([ans[i-1][0] + grid[i][0]]) 
    for j in range(1, m): 
        ans[0].append(ans[0][j-1] + grid[0][j]) 
    for i in range(1, n): 
        for j in range(1, m): 
            ans[i].append(min(ans[i-1][j], ans[i][j-1]) + grid[i][j])
    return ans[n-1][m-1]
```

### Testing:

#### Example 1:

```python
grid = [[6,2,8,9,1],
        [3,4,9,9,0],
        [2,1,1,9,1],
        [3,9,2,5,7]]
minPathSum(grid)
```

Explanation: Because the path 6→3→2→1→1→2→5→7 minimizes the sum: 27

#### Example 2:

```python
grid = [[27,6,0,5,10000],
        [3,20,19,18,30],
        [200,6,4,19,21],
        [7,70,2,500,26]]
minPathSum(grid)
```

Explanation: Because the path 27→6→0→5→18→19→21→26 minimizes the sum: 122
