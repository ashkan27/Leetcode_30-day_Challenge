## Last Stone Weight

### Problem Statement:

We have a collection of stones, each stone has a positive integer weight.

Each turn, we choose the two heaviest stones and smash them together.  Suppose the stones have weights x and y with x <= y.  The result of this smash is:

If x == y, both stones are totally destroyed;
If x != y, the stone of weight x is totally destroyed, and the stone of weight y has new weight y-x.
At the end, there is at most 1 stone left.  Return the weight of this stone (or 0 if there are no stones left.)

### Code:

```python
def lastStoneWeight(stones):
    stones = sorted(stones)
    while len(stones) > 1:
        m = stones.pop()
        m2 = stones.pop()
        d = m - m2
        if d != 0:
            if stones:
                f = False
                for i in range(len(stones)):
                    if stones[i] >= d:
                        f = True
                        break
                if f:
                    stones.insert(i,d)
                else:
                    stones.append(d)
            else:
                return d
    if len(stones) == 0:
        return 0
    else:
        return stones[0]
```

### Testing:

#### Example 1:

```python
stones = [2,7,4,1,8,1]
lastStoneWeight(stones)
```

Input: [2,7,4,1,8,1]

Output: 1

Explanation:

we combine 7 and 8 to get 1 so the array converts to [2,4,1,1,1] then,

we combine 2 and 4 to get 2 so the array converts to [2,1,1,1] then,

we combine 2 and 1 to get 1 so the array converts to [1,1,1] then,

we combine 1 and 1 to get 0 so the array converts to [1] then that's the value of last stone.

#### Example 2:

```python
stones = [5,8,10,7]
lastStoneWeight(stones)
```

Input: [5,8,10,7]

Output: 0

Explanation:

we combine 10 and 8 to get 2 so the array converts to [2,5,7] then,

we combine 7 and 5 to get 2 so the array converts to [2,2] then,

we combine 2 and 2 to get 0 so the array converts to [] then no stones are left.
