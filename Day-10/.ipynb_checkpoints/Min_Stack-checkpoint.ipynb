## Min Stack

### Problem Statement:

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

push(x) -- Push element x onto stack.

pop() -- Removes the element on top of the stack.

top() -- Get the top element.

getMin() -- Retrieve the minimum element in the stack.

### Code:

```python
class MinStack: 
    def __init__(self):
        self.list = []
        
    def push(self, x: int) -> None: 
        if len(self.list)==0:
            self.list.append((x,x))
        else:
            if x <= self.list[-1][0]:
                self.list.append((x,x))
            else:
                self.list.append((self.list[-1][0],x))
                
    def pop(self) -> None:
        self.list.pop()

    def top(self) -> int:
        return self.list[-1][1]

    def getMin(self) -> int:
        return self.list[-1][0]
```

### Testing:

#### Example 1:

```python
ms = MinStack()
result = []
result.append(ms.push(-2))
result.append(ms.push(0))
result.append(ms.push(-3))
result.append(ms.getMin())
result.append(ms.pop())
result.append(ms.top())
result.append(ms.getMin())
result
```

Explanation:

ms = MinStack();

ms.push(-2); // adds -2 to the list. returns None

ms.push(0); // adds 0 to the list. returns None

ms.push(-3); // adds -3 to the list. returns None

ms.getMin(); // returns the minimum -3

ms.pop(); // removes the last element -3 from the list. returns None

ms.top(); // returns the last element 0

ms.getMin(); // returns the minimum -2
