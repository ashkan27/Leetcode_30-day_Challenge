## First Unique Number

### Problem Statement:

You have a queue of integers, you need to retrieve the first unique integer in the queue.

Implement the FirstUnique class:

FirstUnique(int[] nums) Initializes the object with the numbers in the queue.
int showFirstUnique() returns the value of the first unique integer of the queue, and returns -1 if there is no such integer.
void add(int value) insert value to the queue.

### Code:

```python
from collections import OrderedDict
class FirstUnique:
    def __init__(self, nums):
        self.one = OrderedDict()
        self.done = set()
        for i in nums:
            if i in self.one:
                self.one.pop(i)
                self.done.add(i)
            elif i in self.done:
                pass
            else:
                self.one[i] = True
        
    def showFirstUnique(self) -> int:
        if self.one:
            return next(iter(self.one))
        else:
            return -1
        
    def add(self, value: int) -> None:       
        if value in self.one:
            self.one.pop(value)
            self.done.add(value)
        elif value in self.done:
            pass
        else:
            self.one[value] = True
```

### Testing:

#### Example 1:

```python
nums = [2,3,5]
fu = FirstUnique(nums)
result = []
result.append(fu.showFirstUnique())
result.append(fu.add(5))
result.append(fu.showFirstUnique())
result.append(fu.add(2))
result.append(fu.showFirstUnique())
result.append(fu.add(3))
result.append(fu.showFirstUnique())
result
```

Explanation:

nums = [2,3,5]

fu = FirstUnique(nums);

fu.showFirstUnique(); // return 2

fu.add(5);            // the queue is now [2,3,5,5]. return None

fu.showFirstUnique(); // return 3

fu.add(3);            // the queue is now [2,3,5,5,2,3]. return None

fu.showFirstUnique(); // return -1
