## Product of Array Except Self

### Problem Statement:

Given an array nums of n integers where n > 1,  return an array output such that output[i] is equal to the product of all the elements of nums except nums[i].

### Code:

```python
from math import floor
def productExceptSelf(nums):
    n = len(nums)
    m = 1
    z = 0
    for i in nums:
        if i == 0:
            z += 1
        if z == 2:
            break
    if z == 2:
        return [0]*n
    elif z == 1:
        ans = [0]*n
        for i in range(n):
            if nums[i] != 0:
                m *= nums[i]
            else:
                j = i
        ans[j] = m
        return ans
    else:
        ans = []
        for i in nums:
            m *= i
        for i in nums:
            ans.append(floor(m/i))
        return ans
```

### Testing:

#### Example 1:

```python
nums = [4,3,9,3,1]
productExceptSelf(nums)
```

#### Example 2:

```python
nums = [2,0,7,3,4]
productExceptSelf(nums)
```

#### Example 3:

```python
nums = [6,19,0,21,0]
productExceptSelf(nums)
```
