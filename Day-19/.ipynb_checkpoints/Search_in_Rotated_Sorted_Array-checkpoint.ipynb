## Search in Rotated Sorted Array

### Problem Statement:

Suppose an array sorted in ascending order is rotated at some pivot unknown to you beforehand.

(i.e., [0,1,2,4,5,6,7] might become [4,5,6,7,0,1,2]).

You are given a target value to search. If found in the array return its index, otherwise return -1.

You may assume no duplicate exists in the array.

Your algorithm's runtime complexity must be in the order of O(log n).

### Code:

```python
from math import floor
def search(nums, target):
    def binary(nums,target,left,right):
        while (right >= left):
            middle = floor((right+left)/2)
            if nums[middle] == target:
                return middle
            elif nums[middle] > target:
                right = middle-1
            else:
                left = middle+1  
        return -1
    n = len(nums)
    if n == 0:
        return -1
    left = 0
    right = n-1
    f = nums[0]
    if target >= f:
        while(right >= left):
            middle = floor((right+left)/2)
            mv = nums[middle]
            if mv == target:
                return middle
            if mv >= f:
                if mv > target:
                    return binary(nums,target,left,middle-1)
                else:
                    left = middle+1
            else:
                right = middle-1
    else:
        while(right >= left):
            middle = floor((right+left)/2)
            mv = nums[middle]
            if mv == target:
                return middle
            if mv < f:
                if mv < target:
                    return binary(nums,target,middle+1,right)
                else:
                    right = middle-1
            else:
                left = middle+1
    return -1
```

### Testing:

#### Example 1:

```python
nums = [4,5,6,7,0,1,2]
target = 0
search(nums, target)
```

#### Example 2:

```python
nums = [4,5,6,7,0,1,2]
target = 3
search(nums, target)
```

#### Example 3:

```python
nums = [3,6,18,19,20,21,-5,-3,-1,0,1]
target = -3
search(nums, target)
```
