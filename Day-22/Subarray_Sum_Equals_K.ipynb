## Subarray Sum Equals K

### Problem Statement:

Given an array of integers and an integer k, you need to find the total number of continuous subarrays whose sum equals to k.

### Code:

```python
def subarraySum(nums, k):
    s = 0
    dic = {k: 1}
    ans = 0
    for i in nums:
        s += i
        if s in dic:
            ans += dic.get(s)
        if k+s in dic:
            dic[k+s] += 1
        else:
            dic[k+s] = 1
    return ans
```

### Testing:
