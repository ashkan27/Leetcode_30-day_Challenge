## Single Number

### Problem Statement:

Given a non-empty array of integers, every element appears twice except for one. Find that single one.

### Code:

```python
def singleNumber(nums):
    a = nums.pop()
    for i in nums:
        a ^= i
    return a
```

### Testing:

#### Example 1:

```python
nums = [2,2,1]
singleNumber(nums)
```

#### Example 2:

```python
nums = [4,1,2,1,2]
singleNumber(nums)
```

#### Example 3:

```python
nums = [3,5,2,6,4,5,3,1,1,4,9,9,2]
singleNumber(nums)
```
