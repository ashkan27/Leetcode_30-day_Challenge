## Jump Game

### Problem Statement:

Given an array of non-negative integers, you are initially positioned at the first index of the array.

Each element in the array represents your maximum jump length at that position.

Determine if you are able to reach the last index.

### Code:

```python
def canJump(nums):
    del nums[-1]
    bad = [i for i,nums in enumerate(nums) if nums == 0]
    if not bad:
        return True
    goal = bad[-1]+1
    s = 0
    goodindex = []
    goodsum = []
    for b in bad:
        for k in range(1,b+1-s):
            a = b - k
            if nums[a] > k:
                for i in range(s,a+1):
                    goodindex.append(i)
                    goodsum.append(i + nums[i])
                break
        s = b+1
    if 0 not in goodindex:
        return False
    while (goal > 0):
        f = True
        for i in range(len(goodsum)):
            if goodsum[i] >= goal:
                f = False
                break
        if f:
            return False
        goal = goodindex[i]
        goodsum = goodsum[0:i]
    return True
```

### Testing:

#### Example 1:

```python
nums = [2,3,0,1,4]
canJump(nums)
```

Input: [2,3,0,1,4]

Output: True

Explanation: Jump 1 step from index 0 to 1, then 3 steps to the last index.

#### Example 2:

```python
nums = [3,2,1,0,4]
canJump(nums)
```

Input: [3,2,1,0,4]

Output: False

Explanation: You will always arrive at index 3 no matter what. Its maximum jump length is 0, which makes it impossible to reach the last index.

#### Example 3:

```python
nums = [2,0,1,3,0,1,1,0,10,3]
canJump(nums)
```

Input: [2,0,1,3,0,1,1,0,10,3]

Output: False

Explanation: You can never get pass index 7, which makes it impossible to reach the last index.
