## Backspace String Compare

### Problem Statement:

Given two strings S and T, return if they are equal when both are typed into empty text editors. # means a backspace character.

Note that after backspacing an empty text, the text will continue empty.

### Code:

```python
def backspaceCompare(S, T):
    list1 = []
    list2 = []
    for i in S:
        if not list1 and i == "#":
            continue
        if i == "#":
            list1.pop()
        else:
            list1.append(i)
    for i in T:
        if not list2 and i == "#":
            continue
        if i == "#":
            list2.pop()
        else:
            list2.append(i)
    if list1 == list2:
        return True
    else:
        return False
```

### Testing:

#### Example 1:

```python
S = "ab#c"
T = "ad#c"
backspaceCompare(S, T)
```

Input: S = "ab#c", T = "ad#c"


Output: True


Explanation: Both S and T become "ac".

#### Example 2:

```python
S = "a##c"
T = "#a#c"
backspaceCompare(S, T)
```

Input: S = "a##c", T = "#a#c"

Output: True

Explanation: Both S and T become "c".

#### Example 3:

```python
S = "a#c"
T = "b"
backspaceCompare(S, T)
```

Input: S = "a#c", T = "b"

Output: False

Explanation: S becomes "c" while T becomes "b".

#### Example 4:

```python
S = "ab#shj#kaa#n"
T = "###aa#shkan27##"
backspaceCompare(S, T)
```

Input: S = "ab#shj#kaa#n", T = "###aa#shkan27##"

Output: True

Explanation: Both S and T become "ashkan"
