## Group Anagrams

### Problem Statement:

Given an array of strings, group anagrams together.

Note: All inputs will be in lowercase. The order of your output does not matter.

### Code:

```python
def groupAnagrams(strs):
    done = {}
    ans = []
    j = 0
    for i in strs:
        a = "".join(sorted(i))
        if a in done:
            ans[done.get(a)].append(i)
        else:
            done[a] = j
            j += 1
            ans.append([i])
    return ans
```

### Testing:

#### Example 1:

```python
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
groupAnagrams(strs)
```

#### Example 2:

```python
strs = ["bob", "boo", "ooo", "oob", "bbo", "bbb", "obb"]
groupAnagrams(strs)
```
