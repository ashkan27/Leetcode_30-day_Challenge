## Binary Tree Maximum Path Sum

### Problem Statement:

Given a non-empty binary tree, find the maximum path sum.

For this problem, a path is defined as any sequence of nodes from some starting node to any node in the tree along the parent-child connections. The path must contain at least one node and does not need to go through the root

### Code:

```python
# class TreeNode:
#     def __init__(self, value=0, left=None, right=None):
#         self.value = value
#         self.left = left
#         self.right = right

def maxPathSum(root):
    ans = -1 * float("inf")
    def summ(root):
        nonlocal ans
        if not root:
            return 0
        leftsum = max(0, summ(root.left))
        rightsum = max(0, summ(root.right))
        ans = max(ans, root.value + leftsum + rightsum)
        return root.value + max(0, leftsum, rightsum)
    summ(root)
    return ans
```

### Testing:

```python
# To test the code make sure to install binarytree
# pip install binarytree

from binarytree import build
```

#### Example 1:

```python
nodes = [1,2,3] 
tree = build(nodes)
print(tree)
maxPathSum(tree)
```

Explanation:

The path 2 -> 1 -> 3 is the maximum path sum: 6

#### Example 2:

```python
nodes = [-10,9,20,None,None,15,7]
tree = build(nodes)
print(tree)
maxPathSum(tree)
```

Explanation:

The path 15 -> 20 -> 7 is the maximum path sum: 42

#### Example 3:

```python
nodes = [-3, 6, -8, 1, 11, None, 13]
tree = build(nodes)
print(tree)
maxPathSum(tree)
```

Explanation:

The path 11 -> 6 -> -3 -> -8 -> 13 is the maximum path sum: 19
